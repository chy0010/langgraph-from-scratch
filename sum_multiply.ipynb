```python
from typing import List, TypedDict
from  langgraph.graph import StateGraph
import math
```

```python
class AgentState(TypedDict):
    name : str
    values : list[int]
    operation : str 
```

```python
def add_multiply(State:AgentState) -> AgentState :

    if State['operation']=='+':
        State['operation'] = f"Hi {State['name']} your answer is {sum(State['values'])}"
    if State['operation']=='*':
        State['operation'] = f"Hi {State['name']} your answer is {math.prod(State['values'])}"

    return State
```

```python
graph= StateGraph(AgentState)

graph.add_node('addmultiply',add_multiply)
graph.set_entry_point("addmultiply")
graph.set_finish_point("addmultiply")

app=graph.compile()

```

```python
result=app.invoke({"name":"Jack Sparrow","values":[1,2,3,4],"operation":"+"})

print(result["operation"])
```
